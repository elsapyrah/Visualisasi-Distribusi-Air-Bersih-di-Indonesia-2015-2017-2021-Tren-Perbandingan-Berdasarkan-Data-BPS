```python
from sklearn.preprocessing import LabelEncoder#import library LabelEncoder

label_encoder = LabelEncoder()#memanggil library tadi kedalam variabel label_encoder
for i in df:
    df[i] = label_encoder.fit_transform(df[i])#mentransformasikan setiap value kedalam numerik yang terdapat di df_col_category

print('Label Encoded Data: ')#print tulisan

df.head()#menampilkan 5 data teratas
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Scaling
scaler = StandardScaler()
scaler.fit(df)
scaled_df = pd.DataFrame(scaler.transform(df),columns= df.columns )
```

```python
# Membuat matriks korelasi
correlation_matrix = scaled_df.corr()

# Visualisasi matriks korelasi menggunakan heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriks Korelasi")
plt.show()
```

```python
plt.figure(figsize=(12,8)) # Inisiasi ukuran gambar plot yang akan dibuat sebesar 12x8
sns.heatmap(df.corr(), annot=True, linewidths=.5, cmap="Blues") # Membuat visualisasi heatmap berdasarkan nilai korelasi atribut
```
